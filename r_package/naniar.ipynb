```r
library(tidyverse)

airquality %>% 
  bind_shadow() %>% 
  group_by(Ozone_NA) %>% 
  summarise_at(
    "Solar.R",
    c("mean", "sd", "var", "min", "max"),
    na.rm = TRUE
  )
```
