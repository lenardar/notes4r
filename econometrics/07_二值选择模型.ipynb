# Probit模型

```plaintext
library(haven)
womenwk <- read_dta("stata_data/womenwk.dta")
womenwk
```

```plaintext
library(tidyverse)
```

```plaintext
glimpse(womenwk)
```

```plaintext
fit <- lm(work ~ age + married + children + education, data = womenwk)
summary(fit)
```

```plaintext
# 先试用普通OLS
library(sandwich)
robust_se <- vcovHC(fit, type = "HC")

library(lmtest)
coeftest(fit, vcov = robust_se)
```

```plaintext
library(MASS)
```

```plaintext
fit_probit <- glm(
  work ~ age + married + children + education,
  data = womenwk,
  family = binomial(link = "probit"),
  control = glm.control(epsilon=1e-15)
)
summary(fit_probit)
```

```plaintext
library(sandwich)
robust_se_probit <- vcovHC(fit_probit, type = "HC")

library(lmtest)
coeftest(fit_probit, vcov = robust_se_probit)
```

```plaintext
confint(fit_probit)
```

```plaintext
logLik(fit_probit)
```

```plaintext
# 进行LR chi2 检验
# 先结算null model
fit_probit_null <- glm(
  work ~ 1,
  data = womenwk,
  family = binomial(link = "probit")
)

# 计算统计量
chi2 <- -2 * (logLik(fit_probit_null) - logLik(fit_probit))
print(paste("LR chi2 = ", chi2))

# 计算自由度
df <- length(coef(fit_probit)) - length(coef(fit_probit_null))
print(paste("df = ", df))

# 计算p值
p_value <- 1 - pchisq(chi2, df = df)
print(paste("Prob > chi2 = ", p_value))
```

```plaintext
# 计算Pseudo R2
Pseudo_R2 <- (logLik(fit_probit_null) - logLik(fit_probit)) / logLik(fit_probit_null)
print(paste("Pseudo R2 = ", Pseudo_R2))
```

# 稀有事件偏差

```plaintext
library(haven)
nomadic_conquest <- read_dta("stata_data/nomadic_conquest.dta")
nomadic_conquest
```

```plaintext
summary(nomadic_conquest)
```

```plaintext
# 先进行普通logit回归
fit_logit <- glm(
  conquered ~ diff + age + wall + drought1,
  data = nomadic_conquest,
  family = binomial(link = "logit")
)

summary(fit_logit)
```

```plaintext
install.packages("devtools")
devtools::install_github('IQSS/Zelig')
```

```plaintext
library(Zelig)

# 示例：已知总体事件概率为5%，样本中事件比例为3%
tau_value <- 0.05 / (0.05 + 0.97)  # 调整参数计算

# 拟合偏差修正Logistic模型
fit_relogit <- zelig(
  y ~ x1 + x2, 
  model = "relogit", 
  tau = tau_value, 
  bias.correct = TRUE,  # 启用偏差修正
  data = your_data)


summary(fit_relogit)
```

```plaintext
install.packages("brglm2")
```

```plaintext
library(brglm2)

fit_relogit <- glm(
  conquered ~ diff + age + wall + drought1,
  data = nomadic_conquest,
  family = binomial(link = "logit"),
  method = "brglmFit"
)

summary(fit_relogit)
```

```plaintext
# 使用补对数模型
fit_cloglog <- glm(
  conquered ~ diff + age + wall + drought1,
  data = nomadic_conquest,
  family = binomial(link = "cloglog")
)

summary(fit_cloglog)
```

# 工具变量Probit模型

```plaintext
library(haven)
mus14data <- read_dta("stata_data/mus14data.dta")
mus14data
```

```plaintext
print(mus14data, n = Inf)
```

```plaintext
install.packages("ivprobit")
```

```plaintext
library(ivprobit)

fit_ivprobit <- ivprobit(
  ins ~ female + age + age2 + educyear + married + hisp + white + chronic + adl + hstatusg | 
    linc |
    retire,
  data = mus14data
)
```

```plaintext
library(tidyverse)
```

```plaintext
summary(mus14data)
```

```plaintext
# 基于两步法实现
# 删除linc缺失的观测
mus14data <- mus14data |> 
  filter(!is.na(linc))


# 第一步，先得到对于残差的估计
fit_ols <- lm(
  linc ~ retire + # 此为工具变量
    female + age + age2 + educyear + married + hisp + white + chronic + adl + hstatusg,
  data = mus14data
)
mus14data <- mus14data |> 
  mutate(
    res = fit_ols$residuals
  )

# 第二步，加入残差估计，就行probit估计
fit_ivprobit2 <- glm(
  ins ~ linc + res + # 此为内生变量以及一步中的残差
    female + age + age2 + educyear + married + hisp + white + chronic + adl + hstatusg,
  data = mus14data,
  family = binomial(link = "probit")
)

summary(fit_ivprobit2)
```

# 双变量Probit模型

```plaintext
install.packages("mvProbit")
```

```plaintext
library(haven)
school <- read_dta("stata_data/school.dta")
school
```

```plaintext
# 相同解释变量情形
library(mvProbit)

fit_biprobit <- mvProbit(
  cbind(private, vote) ~ loginc + logptax ,
  data = school
)
summary(fit_biprobit)
```

```plaintext
install.packages("GJRM")
```

```plaintext
# 不同解释变量
library(GJRM)

fit_biprobit <- gjrm(
  formula = list(
    private ~ loginc + logptax + school,
    vote ~ loginc + logptax + years + school
  ), 
  data = school,
  model= "B",  # B表示双变量Probit
  margins = c("probit", "probit")
)
summary(fit_biprobit)
```
